```python
data = Path("./versions_example.txt").read_text()
print(data)
```

```python

from typing import Optional
from dataclasses import dataclass

@dataclass
class PatchJob:
    package_id: str
    package_version: Optional[str] # latest if None


def parse_patch_jobs(output):
    for package in data.split("Package name: "):
        package_parts = package.split("Most common compatible versions:")
        if len(package_parts) != 2:
            continue
        package_id = package_parts[0].strip()
        rest = package_parts[1]
        for version in rest.split('\n'):
            version = version.strip().split(' ')[0]
            if version == '':
                continue
            yield PatchJob(package_id=package_id, package_version=None if version == 'Any' else version)

jobs = list(parse_patch_jobs(data))
jobs
```

```python
def download_job_url(job: PatchJob):
    url = f"https://apkpure.com/br/{job.package_id}/"
    if job.package_version is not None:
        url += f"download/{job.package_version}"
    return url

urls = [download_job_url(j) for j in jobs]
urls
```
